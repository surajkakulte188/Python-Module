###**Q.1 Compare and contrast NLTK and spaCy in terms of features, ease of use and performance.**

**→** **NLTK:**
- NLTK (Natural Language Toolkit) is a research-oriented library designed for teaching, linguistics & experimentation.
- It contains many lexical resources such as WordNet & over 50 corpora.
- It offers multiple tokenizers, stemmers, lemmatizers, parsers & classifiers.
- However, NLTK can be slower in performance & is not ideal for production systems.

**SpaCy:**
- SpaCy is an industrial-strength NLP library built for real-world applications.
- It is highly optimized in Cython & provides fast tokenization, part-of-speech tagging, named entity recognition & dependency parsing.
- spaCy includes modern, pretrained neural models & is significantly faster than NLTK.
- Unlike NLTK, spaCy focuses on efficiency & deployment-ready pipelines rather than linguistic variety.

In terms of ease of use, NLTK is more suitable for beginners & academic work, while spaCy is simple, consistent & preferred for production-level systems. Overall, NLTK is best for research & learning whereas spaCy is best for accuracy & performance in real applications.

###**Q.2 What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?**

**→** TextBlob is a high-level NLP library built on top of NLTK & Pattern. It provides a very simple API for performing common NLP tasks without complex code. TextBlob can perform tasks such as sentiment analysis, classification, tokenization, translation, POS tagging & noun phrase extraction.

For sentiment analysis, TextBlob directly returns polarity (negative to positive: -1 to +1) & subjectivity (objective to subjective: 0 to 1). For translation, it integrates with online translation APIs, allowing easy conversion of text from one language to another. Because of its simplicity, TextBlob is widely used by beginners & for rapid prototyping of sentiment & text-processing applications.

###**Q.4 Describe the architecture and functioning of a Recurrent Natural Network (RNN).**

**→** A Recurrent Neural Network (RNN) is a neural model designed for sequential data such as text or time series. Unlike feed-forward networks, RNNs include a recurrent connection that allows information from previous time steps to influence current output.

At each step, the RNN takes an input Xt & the previous hidden state Ht-1. It computes a new hidden state Ht using a nonlinear activation function. This hidden state acts as a memory, enabling the model to capture sequential patterns such as sentence structure.

However, traditional RNNs suffer from vanishing & exploding gradients, making it difficult to learn long-term dependencies. This led to the development of advanced RNN variants like LSTMs & GRUs.

###**Q.5 What is the key difference between LSTM and GRU networks in NLP applications?**

**→**
**LSTM (Long Short-Term Memory):**
- LSTM (Long Short-Term Memory) networks contain three gates: input, forget & output gates.
- They also maintain a separate memory cell to store long-term dependencies.
- This makes LSTMs highly powerful but also computationally heavy.

**GRU (Gated Recurrent Unit):**
- GRU (Gated Recurrent Unit) networks simplify this architecture by using only two gates: update & reset.
- They combine the hidden state & cell state into a single vector.
- GRUs are faster to train & require fewer parameters, yet often achieve performance comparable to LSTMs.

In summary, LSTM handles complex & longer dependencies, whereas GRU is simpler, faster & efficient for smaller datasets or real-time applications.

###**Q.7 Given the sample paragraph below, perform string tokenization and frequency distribution using Python and NLTK: “Natural Language Processing (NLP) is a fascinating field that combines linguistics, computer science and artificial intelligence. It enables machines to understand, interpret and generate human language. Applications of NLP include chatbots, sentiment analysis and machine translation. As technology advances, the role of NLP in modern solutions is becoming increasingly critical.”**

**→**

```python
from nltk.tokenize import RegexpTokenizer
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

text = '''Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science and artificial intelligence. It enables machines to understand,
interpret and generate human language. Applications of NLP include chatbots,
sentiment analysis and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.'''

tokenizer = RegexpTokenizer(r"[A-Za-z]+")
tokens = [t.lower() for t in tokenizer.tokenize(text)]
fd = FreqDist(tokens)

print("Total tokens:", len(tokens))
print("Unique tokens:", len(fd))
print("Top 10 words:", fd.most_common(10))

words, counts = zip(*fd.most_common(10))
plt.figure(figsize=(8,4))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 10 Word Frequencies")
plt.show()
```

###**Q.8 Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).**
Dataset texts = [

“I love this project”, #Positive

“This is an amazing experience”, #Positive

“I hate waiting in line”, #Negative

“This is the worst service”, #Negative

“Absolutely fantastic!” #Positive

]

labels = [1, 1, 0, 0, 1]
###**Preprocess the text, tokenize it, pad sequences and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.**

**→**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
"I love this project",
"This is an amazing experience",
"I hate waiting in line",
"This is the worst service",
"Absolutely fantastic!"
]
labels = np.array([1, 1, 0, 0, 1])

tok = Tokenizer(oov_token="")
tok.fit_on_texts(texts)
seqs = tok.texts_to_sequences(texts)
max_len = max(len(s) for s in seqs)
X = pad_sequences(seqs, maxlen=max_len, padding="post")
vocab_size = len(tok.word_index) + 1

print("Vocab size:", vocab_size)
print("Max length:", max_len)
print("Padded sequences:\n", X)

model = Sequential([
Embedding(input_dim=vocab_size, output_dim=16, input_length=max_len),
LSTM(16),
Dense(1, activation="sigmoid")
])
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()
model.fit(X, labels, epochs=40, verbose=0)
loss, acc = model.evaluate(X, labels, verbose=0)
print("Train Accuracy:", acc)

pred = model.predict(X, verbose=0).ravel()
print("Predicted probabilities:", np.round(pred, 3))
print("Predicted labels:", (pred >= 0.5).astype(int).tolist())
```

###**Q.9 Using spaCy, build a simple NLP pipeline that includes tokenization, lemmatization and entity recognition. Use the following paragraph as your dataset:**
“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the development of India's atomic energy program. He was the founding director of the Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing the Atomic Energy Commission of India.”
###**Write a Python program that processes this text using spaCy, then prints tokens, their lemmas and any named entities found.**


**→**

```python
import spacy
from spacy.cli import download

model_name = "en_core_web_sm"
try:
    nlp = spacy.load(model_name)
except Exception:
    download(model_name)
    nlp = spacy.load(model_name)

text = (
    "Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the "
    "development of India's atomic energy program. He was the founding director of the Tata "
    "Institute of Fundamental Research (TIFR) and was instrumental in establishing the "
    "Atomic Energy Commission of India."
)

doc = nlp(text)

print("TOKENS + LEMMAS:")
for t in doc:
    print(f"{t.text:20s} -> {t.lemma_}")

print("\nNAMED ENTITIES:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

###**Q.10 You are working on a chatbot for a mental health platform. Explain how you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford NLP to understand and respond to user input effectively. Detail your architecture, data preprocessing pipeline and any ethical considerations.**

**→** **Architecture:**
1. Embedding Layer → LSTM/GRU → Dense Output
2. Classification tasks: emotion detection, intent classification
3. spaCy handles: Tokenization, Lemmatization & Named Entity Recognition.

**NLP Pipeline:**
1. Clean user messages
2. Lemmatize using spaCy
3. Convert to sequences
4. Classify using LSTM/GRU
5. Generate empathetic response templates.

**Ethical Consideration:**
1. No harmful responses
2. Crisis detection (“self-harm”, “suicide”) must trigger escalation
3. Users' privacy & confidentiality must be ensured
4. Avoid hallucination; always provide disclaimers.

**Sample Code:**

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense

train_texts = [
"hi", "hello there", "good morning",
"i feel anxious", "i am stressed", "i am worried",
"thank you", "thanks a lot",
"bye", "goodbye",
"i want to hurt myself", "i can't go on"
]
train_labels = [
0,0,0, # greeting
1,1,1, # anxiety/stress
2,2, # thanks
3,3, # goodbye
4,4 # crisis-risk (toy)
]
label_names = {0:"greeting", 1:"anxiety", 2:"thanks", 3:"goodbye", 4:"crisis"}

tok2 = Tokenizer(oov_token="")
tok2.fit_on_texts(train_texts)
seqs = tok2.texts_to_sequences(train_texts)
max_len = max(len(s) for s in seqs)

X = pad_sequences(seqs, maxlen=max_len, padding="post")
y = np.array(train_labels)

vocab_size = len(tok2.word_index) + 1

model2 = Sequential([
Embedding(vocab_size, 16, input_length=max_len),
GRU(16),
Dense(len(label_names), activation="softmax")
])
model2.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model2.fit(X, y, epochs=60, verbose=0)

def safe_reply(intent):
    if intent == "greeting":
        return "Hi! I’m here to listen. How are you feeling today?"
    if intent == "anxiety":
        return "I’m sorry you’re feeling stressed. Want to share what’s bothering you?"
    if intent == "thanks":
        return "You’re welcome. I’m glad I could help."
    if intent == "goodbye":
        return "Take care. If you need to talk again, I’m here."
    if intent == "crisis":
        return ("I’m really sorry you’re feeling this way. I’m not a replacement for professional help. "
        "If you feel in danger right now, please contact your local emergency number or reach out to a trusted person immediately.")
    return "I’m here to help."

test_msgs = ["hello", "i am very stressed", "thanks", "bye", "i can't go on anymore"]
test_pad = pad_sequences(tok2.texts_to_sequences(test_msgs), maxlen=max_len, padding="post")

pred = model2.predict(test_pad, verbose=0).argmax(axis=1)

for msg, pid in zip(test_msgs, pred):
    intent = label_names[int(pid)]
    print("User:", msg)
    print("Predicted intent:", intent)
    print("Bot:", safe_reply(intent))
    print("-"*50)
```
